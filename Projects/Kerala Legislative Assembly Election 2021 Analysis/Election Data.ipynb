# Data Visualization

```python
from matplotlib import pyplot as plt

x = np.array(LDF_percentage_total)
y = np.array(alappuzha_total)

plt.subplot(1,2,1)
plt.plot(x,y,marker="s",color='red',ms=20)
plt.title('LDF')

plt.subplots_adjust(right=2)

x = np.array(UDF_percentage_total)
y = np.array(alappuzha_total)

plt.subplot(1,2,2)
plt.plot(x,y,marker="d",color='blue',ms=20)
plt.title('UDF')

plt.suptitle('LDF vs UDF in Alappuzha Constituency')
plt.show()
```
